## Perform backtest on 3 years data and provide complete analysis of the strategy using pyfolio.

```python
import yfinance as yf
import pandas as pd
import talib
import pyfolio as pf

def get_top_performers(symbols, start_date, end_date):
    returns = []
    
    # Fetch historical data for each stock
    for symbol in symbols:
        # Fetch data for the current stock
        stock = yf.Ticker(symbol)
        stock_data = stock.history(start=start_date, end=end_date)
        
        # Calculate the weekly returns
        weekly_returns = stock_data['Close'].pct_change().resample('W').sum()
        avg_weekly_return = weekly_returns.mean()
        
        # Store the stock symbol and average weekly return
        returns.append((symbol, avg_weekly_return))
    
    # Create a DataFrame from the returns list
    returns_df = pd.DataFrame(returns, columns=['Symbol', 'Average Weekly Return'])
    
    # Sort the DataFrame by average weekly return in descending order
    returns_df = returns_df.sort_values(by='Average Weekly Return', ascending=False)
    
    # Get the top 5 performers
    top_performers = returns_df.head(5)
    
    return top_performers

def apply_stop_loss(entry_price, atr):
    stop_loss = entry_price - (2 * atr)
    return stop_loss

def rebalance_portfolio(top_performers):
    # Calculate the allocation for each stock
    allocation = 1 / len(top_performers)
    
    trades = []
    
    # Calculate the number of shares to buy and apply stop loss for each stock
    for symbol in top_performers['Symbol']:
        stock = yf.Ticker(symbol)
        last_close_price = stock.history(period='1d').iloc[-1]['Close']
        
        # Calculate ATR
        high = stock.history(period='1y')['High']
        low = stock.history(period='1y')['Low']
        close = stock.history(period='1y')['Close']
        atr = talib.ATR(high, low, close, timeperiod=14).iloc[-1]
        
        shares_to_buy = allocation / last_close_price
        stop_loss = apply_stop_loss(last_close_price, atr)
        
        # Store the trade details
        trade = {
            'Date': pd.Timestamp.today(),
            'Symbol': symbol,
            'Shares': shares_to_buy,
            'Stop Loss': stop_loss,
            'amount': shares_to_buy * last_close_price,
            'price': last_close_price  # Add the 'price' column
        }
        trades.append(trade)
    
    # Create a DataFrame from the trades list
    trades_df = pd.DataFrame(trades)
    trades_df.set_index('Date', inplace=True)
    
    return trades_df



# Define the list of stock symbols
symbols = [
    'ABB.NS',
    'ACC.NS',
    'AUBANK.NS',
    'AARTIIND.NS',
    'ABBOTINDIA.NS',
    'ADANIENT.NS',
    'ADANIPORTS.NS',
    'ABCAPITAL.NS',
    'ABFRL.NS',
    'ALKEM.NS',
    'AMBUJACEM.NS',
    'APOLLOHOSP.NS',
    'APOLLOTYRE.NS',
    'ASHOKLEY.NS',
    'ASIANPAINT.NS',
    'ASTRAL.NS',
    'ATUL.NS',
    'AUROPHARMA.NS',
    'AXISBANK.NS',
    'BAJAJ-AUTO.NS',
    'BAJFINANCE.NS',
    'BAJAJFINSV.NS',
    'BALKRISIND.NS',
    'BALRAMCHIN.NS',
    'BANDHANBNK.NS',
    'BANKBARODA.NS',
    'BATAINDIA.NS',
    'BERGEPAINT.NS',
    'BEL.NS',
    'BHARATFORG.NS',
    'BHEL.NS',
    'BPCL.NS',
    'BHARTIARTL.NS',
    'BIOCON.NS',
    'BSOFT.NS',
    'BOSCHLTD.NS',
    'BRITANNIA.NS',
    'CANFINHOME.NS',
    'CANBK.NS',
    'CHOLAFIN.NS',
    'CIPLA.NS',
    'CUB.NS',
    'COALINDIA.NS',
    'COFORGE.NS',
    'COLPAL.NS',
    'CONCOR.NS',
    'COROMANDEL.NS',
    'CROMPTON.NS',
    'CUMMINSIND.NS',
    'DLF.NS',
    'DABUR.NS',
    'DALBHARAT.NS',
    'DEEPAKNTR.NS',
    'DELTACORP.NS',
    'DIVISLAB.NS',
    'DIXON.NS',
    'LALPATHLAB.NS',
    'DRREDDY.NS',
    'EICHERMOT.NS',
    'ESCORTS.NS',
    'EXIDEIND.NS'
]

# Specify the start and end dates
start_date = '2022-06-01'
end_date = '2023-06-01'

# Get the top 5 performers based on weekly returns
top_performers = get_top_performers(symbols, start_date, end_date)

# Rebalance the portfolio
trades_df = rebalance_portfolio(top_performers)

# Perform backtest and analyze using Pyfolio
returns = pd.DataFrame()
for symbol in trades_df['Symbol']:
    stock_returns = yf.Ticker(symbol).history(start=start_date, end=end_date)['Close'].pct_change()
    returns = pd.concat([returns, stock_returns], axis=1)
returns.columns = trades_df['Symbol']
returns.dropna(inplace=True)

# Convert returns DataFrame to Pyfolio compatible format
returns_pyfolio = returns.copy()
returns_pyfolio.index = pd.to_datetime(returns_pyfolio.index)

# Create positions DataFrame
positions = pd.DataFrame(index=returns_pyfolio.index)

# Create transactions DataFrame
transactions = trades_df.copy()
transactions['Symbol'] = transactions['Symbol'].apply(lambda x: x.replace('.NS', ''))

# Specify the out-of-sample date
oos_date = pd.Timestamp.today()

# Create sector mappings dictionary (if available)
sector_map = {}

# Create the tear sheet
pf.create_full_tear_sheet(
    returns=returns_pyfolio,
    positions=positions,
    transactions=transactions,
    live_start_date=oos_date,
    slippage=0.1,
    sector_mappings=sector_map
)
```
